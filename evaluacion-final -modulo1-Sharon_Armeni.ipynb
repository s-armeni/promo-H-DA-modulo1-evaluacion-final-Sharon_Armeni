```python
class TiendaOnline:
    def __init__(self,inventario = [],clientes = {},ventas_totales = 0.0):
        self.inventario = inventario # una lista de diccionarios para almacenar los productos del inventario: [{'nombre': ' ', 'precio': , 'cantidad': }].Al principio deberá ser una lista vacia.
        self.clientes = clientes # un diccionario para llevar un registro de los clientes:{'Cliente1': {'email': 'cliente1@email.com', 'compras': []}
        self.ventas_totales = ventas_totales # un float inicialmente en 0.0 para llevar un registro de las ventas de la tienda.

    def agregar_producto(self, nombre, precio, cantidad):
        
        agregado = False
        for producto in self.inventario: # producto en este caso es el diccionario.
            for campo,valor in producto.items():
                if campo == "Nombre" and valor == nombre:
                    producto["Cantidad"] = cantidad 
                    print(f"Se ha actualizado la cantidad del producto {nombre} y la nueva cantidad es {cantidad}")
                    agregado = True
                    break
        if agregado == False:
            nuevo_producto =  {"Nombre": nombre, "Precio": precio, "Cantidad": cantidad}
            self.inventario.append(nuevo_producto)
            print("Se ha agregado el nuevo producto al inventario")

    def ver_inventario(self):
    
        for producto in self.inventario: 
            print(str(producto)[1:-1])

    def buscar_producto(self, nombre):
    
        for producto in self.inventario:
            for campo,valor in producto.items():
                if campo == "Nombre" and valor == nombre:
                    print(str(producto)[1:-1])
    
    def actualizar_stock(self, nombre, cantidad = input):
        cantidad = input("Ingrese la cantidad del producto: ")

        actualizado = False
        for producto in self.inventario:
            for campo,valor in producto.items():
                if campo == "Nombre":
                    if valor == nombre:
                        producto["Cantidad"] = cantidad
                        print(f"El stock del producto {nombre} ha sido actualizado. La nueva cantidad es {cantidad}")
                        actualizado = True
                        break                
        if actualizado == False:
            print(f"El producto {nombre} no está en el stock")

    def eliminar_producto(self, nombre):    
        eliminado = False
        for producto in self.inventario:
            for campo,valor in producto.items():
                if campo == "Nombre":
                    if valor == nombre:
                        self.inventario.remove(producto)
                        print(f"El producto {nombre} ha sido eliminado")
                        eliminado = True
                        break
        if eliminado == False:
            print(f"El producto {nombre} no ha sido encontrado")
            

    def calcular_valor_inventario(self):
        
        valor_inventario_total = 0
        for producto in self.inventario:
            precio = producto ["Precio"]
            cantidad = producto ["Cantidad"]
            valor_inventario_total += precio * cantidad
        print(f"El valor total del inventario es: {valor_inventario_total}")

    def realizar_compra(self):
    
        compra_finalizada = False
        while compra_finalizada != True:
            carrito = []
            total_compra = 0
            continuar_compra = input("Quiere añadir algún producto a su carrito: si/no?")
            print(f"Este es el inventario:{self.inventario}")
            if continuar_compra == "no":
                compra_finalizada = True
                print("El proceso de compra ha finalizado")
            elif continuar_compra == "si":
                producto_deseado = input ("Ingrese el nombre del producto que desea comprar: ")
                for producto in self.inventario:
                    for campo,valor in producto.items():
                        if campo == "Nombre":
                            if valor == producto_deseado:
                                carrito.append(producto_deseado)
                                print(f"El producto {producto_deseado} ha sido añadido a su carrito")
                                producto["Cantidad"] -= 1
                                total_compra = total_compra + producto["Precio"]
                                print("El carrito contiene:", carrito)
                                print("El total de la compra es:",total_compra)
                        
       
mi_tiendaonline = TiendaOnline()   # Instancio la clase

mi_tiendaonline.inventario

mi_tiendaonline.agregar_producto("Pantalon", 40, 8)
mi_tiendaonline.agregar_producto("Camiseta", 30, 9)
mi_tiendaonline.agregar_producto("Vestido", 50, 10)

mi_tiendaonline.inventario #imprimo para ver la diferencia

mi_tiendaonline.ver_inventario()

mi_tiendaonline.inventario

mi_tiendaonline.buscar_producto("Vestido")

# Descomentar estas lineas para testear el metodo actualizar_stock
# mi_tiendaonline.actualizar_stock("Pantalon")
# mi_tiendaonline.inventario

mi_tiendaonline.eliminar_producto("Falda")

mi_tiendaonline.eliminar_producto("Pantalon")

mi_tiendaonline.inventario

mi_tiendaonline.ver_inventario()

mi_tiendaonline.realizar_compra()

mi_tiendaonline.ver_inventario()

mi_tiendaonline.calcular_valor_inventario()
```
